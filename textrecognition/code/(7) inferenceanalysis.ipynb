# 추론 결과 분석

```python
import pandas as pd

df = pd.read_csv('inferenceresult.csv')
df
```

```python
# 틀린 텍스트 이미지 정보 모으기
dic_error = {'file_name':[], "original_label":[], "generated_text":[] }
for idx, data in df.iterrows():
    if data['original_label'] != data['generated_text'] :
        dic_error['file_name'].append(data['file_name'])
        dic_error['original_label'].append(data['original_label'])
        dic_error['generated_text'].append(data['generated_text'])
dic_error
```

```python
# 틀린 이미지만 따로 폴더에 모으기
from PIL import Image
import shutil

for img in dic_error['file_name']:
    from_file_path = 'data/croppedimages/' + img  # 복사할 파일
    to_file_path = 'inferenceanalysis/' + img # 복사 위치 및 파일 이름 지정
    shutil.copyfile(from_file_path, to_file_path) 
```

```python
# 이번엔, 맞은 이미지만 폴더에 모으기
dic_right = {'file_name':[], "original_label":[], "generated_text":[] }
for idx, data in df.iterrows():
    if data['original_label'] == data['generated_text'] :
        dic_right['file_name'].append(data['file_name'])
        dic_right['original_label'].append(data['original_label'])
        dic_right ['generated_text'].append(data['generated_text'])
dic_right

for img in dic_right['file_name']:
    from_file_path = 'data/croppedimages/' + img  # 복사할 파일
    to_file_path = 'inferenceanalysis/right/' + img # 복사 위치 및 파일 이름 지정
    shutil.copyfile(from_file_path, to_file_path) 
```

```python
# 실제 라벨과 예측한 텍스트 확인
import csv

df_error = pd.DataFrame(dic_error)
for idx, data in df_error.iterrows():
    with open('inf.csv', 'a') as f:
        info = [{'file_name': data['file_name'], 'original_label': data['original_label'], 'generated_text':  data['generated_text'] }]
        writer = csv.DictWriter(f, fieldnames=info[0].keys())
        writer.writerows(info)
```

```python
df_right = pd.DataFrame(dic_right)
for idx, data in df_right.iterrows():
    with open('right.csv', 'a') as f:
        info = [{'file_name': data['file_name'], 'original_label': data['original_label'], 'generated_text':  data['generated_text'] }]
        writer = csv.DictWriter(f, fieldnames=info[0].keys()
        writer.writerows(info)
```
